### While Loop
    A while loop is used to execute a block of statements repeatedly until a given condition is satisfied. When the condition becomes false, the line immediately after the loop in the program is executed.
    
    Python While Loop is used to execute a block of statements repeatedly until a given condition is satisfied. When the condition becomes false, the line immediately after the loop in the program is executed.



```python
count = 0

while count < 3:
    print(f'Hey Rajkumar the count is {count}')
    count += 1
```

```python
# Infinite While Loop:- If we want a block of code to execute infinite number of times then we can use the while loop in Python to do so.

while (True):
    print("Hello Raj", end=" ")
    break
```

```python
# Infinite while Loop:- the value of the condition is always True. Therefore, the body of the loop is run infinite times until the memory is full.

age = 27

while age > 19:
    print('Infinite time age')
    break
```

```python
# while loop with continue statement
# Prints all letters except 'a' and 'u'

q = 'Raj Kumar Malyala'
count = 0

while count < len(q):
    # print(q[count])
    if q[count] == 'a' or q[count] == 'u':
        count += 1
        continue
    print(q[count])
    count += 1
```

```python
# while loop with break statement
i = 0
a = 'geeksforgeeks'
while i < len(a):
    if a[i] == 'e' or a[i] == 's':
        i += 1
        print(i)
        break
    print(a[i])
    i += 1
```

```python
# while loop with pass statement:-  pass statement to write empty loops. Pass is also used for empty control statements, functions, and classes.
a = 'geeksforgeeks'
i = 0
while i < len(a):
    i += 1
    pass
print('Value of i :', i)
```

```python
"""
while loop with else

while loop executes the block until a condition is satisfied. When the condition becomes false, the statement immediately after the loop is executed. 
The else clause is only executed when your while condition becomes false.
Note: The else block just after for/while is executed only when the loop is NOT terminated by a break statement.

"""

i = 0
while i < 4:
    i += 1
    print(i)
else:  # Executed because no break in for
    print("No Break\n")

i = 0
while i < 4:
    i += 1
    print(i)
    continue
else:  # Not executed as there is a break
    print("No Break")
```
