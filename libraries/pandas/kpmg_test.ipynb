`2. Which of the following tokenization techniques are valid with respect to NLP?`

1. Data tokenization and result tokenization
2. Whitespace / unigram tokenization and regular expression tokenization
3. Output tokenization and string tokenization
4. Logical tokenization and syntax tokenization

Answer:- 2. Whitespace / unigram tokenization and regular expression tokenization

Whitespace / Unigram Tokenization

- Splits text based on spaces.

    Example:
    "I love NLP" ‚Üí ["I", "love", "NLP"]

- This is one of the most basic and commonly used NLP tokenization techniques.

Regular Expression Tokenization

- Uses regex rules to split text based on patterns (punctuation, numbers, symbols).

    Example:
    "Hello, world!" ‚Üí ["Hello", "world"]

Widely used in real-world NLP preprocessing.

‚úîÔ∏è Both are standard, valid, and widely accepted NLP tokenization techniques.

`3. What does the term ‚Äúoverfitting‚Äù mean in the context of training an autoencoder?`

1. The model fails to learn any patterns from the data
2. The model performs well on the training data but poorly on new, unseen data
3. The model cannot compress data into a lower-dimensional space
4. The model only learns linear relationships

Answer: 2. The model performs well on the training data but poorly on new, unseen data

Overfitting is a core machine learning concept, and it applies to autoencoders as well.

- An autoencoder is trained to reconstruct its input.

When it overfits, it:

- Memorizes the training data
- Learns noise or very specific patterns
- Fails to generalize to unseen data

üìå Key idea:

- High performance on training data + poor performance on validation/test data = Overfitting
- This definition is universal across ML models, including autoencoders.
